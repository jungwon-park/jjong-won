```python
f = open("population.csv","rt")
one = 0
two = 0
three = 0
four = 0
five = 0
six = 0
seven = 0
while True:
    line = f.readline()
    if not line:
        break    
    if line[2] == '0':
        one += 1
    elif line[2] == '1':
        two += 1
    elif line[2] == '2':
        three += 1
    elif line[2] == '3':
        four += 1
    elif line[2] == '4':
        five += 1 
    elif line[2] == '5':
        six += 1 
    else:
        seven += 1 
print(f"10대 미만:{one}")
print(f"10대:{two}")    
print(f"20대:{three}")
print(f"30대:{four}")
print(f"40대:{five}")
print(f"50대:{six}")
print(f"60대 이상:{seven}")
f.close()
```
